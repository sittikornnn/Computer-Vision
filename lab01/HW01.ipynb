# assignments [4/5]

```python
import math

def euclidean_distance(d1,d2):
    squared_distances = [(p1 - p2) ** 2 for p1, p2 in zip(d1, d2)]
    sum_of_squares = sum(squared_distances)
    distance = math.sqrt(sum_of_squares)
    return distance
```

```python
d_a_b = euclidean_distance(a,b)
d_a_b
```

```python
d_a_c = euclidean_distance(a,c)
d_a_c
```

```python
d_a_d = euclidean_distance(a,d)
d_a_d
```

```python
d_b_c = euclidean_distance(b,c)
d_b_c
```

```python
d_b_d = euclidean_distance(b,d)
d_b_d
```

```python
d_c_d = euclidean_distance(c,d)
d_c_d
```

```python
M = min(d_a_b,
           d_a_c,
           d_a_d,
           d_b_c,
           d_b_d,
           d_c_d)
```
